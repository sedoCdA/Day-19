```
def mean(data):
    total = 0
    count = 0
    for x in data:
        total += x
        count += 1
    return total / count


def median(data):
    s = sorted(data)
    n = len(s)
    mid = n // 2

    if n % 2 == 1:
        return s[mid]
    else:
        return (s[mid - 1] + s[mid]) / 2


def variance(data):
    m = mean(data)
    total = 0
    count = 0

    for x in data:
        total += (x - m) ** 2
        count += 1

    return total / count


data = [12, 23, 3, 34, 45, 56, 5]

print("Mean:", mean(data))
print("Median:", median(data))
print("Variance:", variance(data))
```

```
from functools import reduce

def mean(data):
    return reduce(lambda x, y: x + y, data) / len(data)

def variance(data):
    m = mean(data)
    squared_diff = list(map(lambda x: (x - m) ** 2, data))
    return reduce(lambda x, y: x + y, squared_diff) / len(data)

print("Mean:", mean(data))
print("Median:", median(data))
print("Variance:", variance(data))
```

```
students = [50, 60, 70, 80, 90]

avg = mean(students)

above_avg = list(filter(lambda x: x > avg, students))

grades = ["A" if x >= 80 else "B" if x >= 60 else "C" for x in students]

print(above_avg)
print(grades)
```

```
import numpy as np

def std_dev(data):
    return variance(data) ** 0.5

data = [10, 20, 30, 40, 50]

print(std_dev(data))
print(np.std(data))
```

```
A = [12,23,3,34,45,56,5]
B = [12,1,3,1,1,2,3,4,5,3,4]

mean_A = mean(A)
mean_B = mean(B)

var_A = variance(A)
var_B = variance(B)

diff = mean_A - mean_B

print(mean_A, var_A)
print(mean_B, var_B)
print("Difference:", diff)
```

```
import math

def t_test(sample1, sample2):
    m1, m2 = mean(sample1), mean(sample2)
    v1, v2 = variance(sample1), variance(sample2)

    n1, n2 = len(sample1), len(sample2)

    # t-statistic
    t = (m1 - m2) / math.sqrt((v1/n1) + (v2/n2))

    return(t)

A = [12,23,3,34,45,56,5]
B = [12,1,3,1,1,2,3,4,5,3,4]

print(t_test(A, B))
```

```
t_value = t_test(A, B)

if abs(t_value) > 2:
    print("Significantly different")
else:
    print("Not significantly different")
```
