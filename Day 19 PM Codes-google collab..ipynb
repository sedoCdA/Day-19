```
import numpy as np

data = np.random.normal(50, 10, 1000)
```

```
mean = np.mean(data)
variance = np.var(data)
std_dev = np.std(data)

print(mean, variance, std_dev)
```

```
import matplotlib.pyplot as plt

plt.hist(data, bins=30)
plt.title("Normal Distribution")
plt.show()
```

```
def bernoulli_pmf(x, p):
    if x == 1:
        return p
    elif x == 0:
        return 1 - p

# Test
p = 0.6
print(bernoulli_pmf(1, p))
print(bernoulli_pmf(0, p))

# Verify total probability
print(bernoulli_pmf(1, p) + bernoulli_pmf(0, p))
```

```
import math

def normal_pdf(x, mean, std):
    return (1 / (std * math.sqrt(2 * math.pi))) * math.exp(-((x - mean) ** 2) / (2 * std ** 2))

# Test
print(normal_pdf(50, 50, 10))
```

```
import statistics

data = [10, 20, 20, 30, 40]

mean = statistics.mean(data)
median = statistics.median(data)
mode = statistics.mode(data)

print(mean, median, mode)
```

```
if mean > median:
    print("Right skewed")
elif mean < median:
    print("Left skewed")
else:
    print("Symmetric")
```

```
import random

heads = 0
trials = 1000

for _ in range(trials):
    if random.choice(["H", "T"]) == "H":
        heads += 1

probability = heads / trials

print("Experimental:", probability)
print("Theoretical:", 0.5)
```

```
normal = np.random.normal(0, 1, 1000)
uniform = np.random.uniform(0, 1, 1000)

plt.hist(normal, bins=30)
plt.title("Normal Distribution")
plt.show()

plt.hist(uniform, bins=30)
plt.title("Uniform Distribution")
plt.show()
```
